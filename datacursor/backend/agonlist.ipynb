```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
from optparse import Values
from os import name


class FirstAgentState(TypedDict):
    values: list[int]
    name:str
    result:str
def process_list(state:FirstAgentState)->FirstAgentState:
    """ This function handles multiple different inputs"""
    state['result']=f"Heloo {state['name']}, the sum of the list {state['values']} is {sum(state['values'])}"
    return state
graph=StateGraph(FirstAgentState)
graph.add_node("process_list",process_list)
graph.set_entry_point("process_list")
graph.set_finish_point("process_list")
app=graph.compile()
ans=app.invoke({'values':[1,2,4,5],'name':"Raghavendra"})
```

```python
print(ans['result'])
```
